### Minimize the 2D Rosenbrock function (given below) using the standard gradient descent, i.e where $\alpha$ (step size) is constant. (Don't forget to check the solution for convergence)
### Plot the Rosenbrock function on a contour graph ($X_{min}=Y_{min}=-2, X_{max}=Y_{max}=2$)
### On the same graph plot the path taken  by the gradient descent algorithm
### To experiment, start the algorithm at four different points; $\mathbf{x_1}=[-1.5,-1.5]$, $\mathbf{x_2}=[-1.9,1.9]$, $\mathbf{x_3}=[1.9,1.9]$,  $\mathbf{x_4}=[1.9,-1.9]$

```python
def rosenbrock(x, a=1, b=5):
    """
    Rosenbrock's function is a 2D uni-modal function
    This implementation follows the definition in Kochenderfer & Wheeler, 2019
    x : x is a numpy array of dimensions [2,m], where m is the number of 2D points
    """
    return (a-x[0])**2 + b*(x[1] - x[0]**2)**2
```

```python
#gradient descent algorithm returning all points visited
def gradient_descent(f, x0, alpha, max_iter=1000, tol=1e-6):
    """
    Gradient descent algorithm
    f : function to minimize
    x0 : initial point
    alpha : learning rate
    max_iter : maximum number of iterations
    tol : tolerance for convergence
    """
    x = x0
    x_list = [x0]
    for i in range(max_iter):
        #compute gradient
        grad = np.zeros(x.shape)
        for j in range(x.shape[0]):
            x1 = x.copy()
            x1[j] = x1[j] + 1e-6
            grad[j] = (f(x1) - f(x))/1e-6
        #update x
        
        x = x - alpha*grad
      

        x_list.append(x)
        if np.linalg.norm(grad) < tol:
            break
    return np.array(x_list)
      

    
```

```python
x0 = np.array([-1.5, -1.5])
x1 = np.array([-1.9, 1.9])
x2 = np.array([1.9, 1.9])
x3 = np.array([1.9, -1.9])
x_list = gradient_descent(rosenbrock, x0, alpha=0.01, max_iter=1000, tol=1e-6)
x_list1 = gradient_descent(rosenbrock, x1, alpha=0.01, max_iter=1000, tol=1e-6)
x_list2 = gradient_descent(rosenbrock, x2, alpha=0.01, max_iter=1000, tol=1e-6)
x_list3 = gradient_descent(rosenbrock, x3, alpha=0.01, max_iter=1000, tol=1e-6)
```

### The gradient descent can be  accelerated by computing the optimal step size, i.e $\alpha_{opt}$
###  $\alpha_{opt}$ is obtained by minimising the univariate function, $f(\mathbf{x}+\alpha \mathbf{d})$ over $\alpha$,  where $\mathbf{x}$ is the input vector of the loss function $f$, and $\mathbf{d}$ is the directional derivative at the point $\mathbf{x}$.
### Use the golden search technique to compute the optimal step size in the standard gradient descent  when minimizing the 2D Rosenbrock function and include in your answer a graph of the Rosenbrock function and a plot of the path taken by the gradient descent algorithm with optimal step size.
# [40 marks]




```python
def golden_search(f, a, b, e=1e-6, max_itr=100):
    """
    Golden search algorithm for univariate function
    f : univariate function
    a : lower bound of interval
    b : upper bound of interval
    e : tolerance
    max_itr : maximum number of iterations
    returns:
    x : point at which minimum is found
    n : number of iterations
    """
    phi = (1 + np.sqrt(5)) / 2
    c = b - (b - a) / phi
    d = a + (b - a) / phi
    fc = f(c)
    fd = f(d)
    n = 0
    while (b - a) > e and n < max_itr:
        if fc < fd:
            b = d
            d = c
            c = b - (b - a) / phi
            fd = fc
            fc = f(c)
        else:
            a = c
            c = d
            d = a + (b - a) / phi
            fc = fd
            fd = f(d)
        n += 1
    x = (a + b) / 2
    return x, n
```

```python
#gradient descent on rosenbrock function with optimized step size using golden search
def gradient_descent_opt(f, x0, max_iter=1000, tol=1e-6):
    """
    Gradient descent algorithm
    f : function to minimize
    x0 : initial point
    max_iter : maximum number of iterations
    tol : tolerance for convergence
    """
    x = x0
    x_list = [x0]
    for i in range(max_iter):
        #compute gradient
        grad = np.zeros(x.shape)
        for j in range(x.shape[0]):
            x1 = x.copy()
            x1[j] = x1[j] + 1e-6
            grad[j] = (f(x1) - f(x))/1e-6
        #update x
        alpha = golden_search(lambda alpha: f(x - alpha*grad), 0, 1)[0]
        x = x - alpha*grad
        x_list.append(x)
        if np.linalg.norm(grad) < tol:
            break
    return np.array(x_list)
    
```

```python
def cce2503_2022_loss_function(z):
    """
    z : vector of length 5
    returns:
    scalar value of function at point z
    """
    assert z.flatten().ravel().shape[0]==5, "z is not of length 5"
    z=z.reshape(1,5)
    X,y=pickle.load(open("numbers.pck","rb"))
    p=np.matmul(z,X.T)
    e=((y.ravel()-p.ravel())**2.0).mean()
    return e
    
```

### Code the coordinate descent algorithm to minimise the loss function
### What is $\mathbf{z}$ that minimises the loss function, up to a volume of uncertainty that is equal to 1e-15?
### How many function calls are required to reach the minimum with a volume of uncertainty that is equal to 1e-15?
# [40 marks]

```python
class function_call_count:
    def __init__(self, name, count):
        self.name = name
        self.count = count

    def display_count(self): 
        print("Function call count is ", self.count)
        
    def reset_count(self): 
        self.count = 0
    
    def call_func(self,func):  

        def wrapper(*args, **kwargs):
            self.count += 1  
            result = func(*args, **kwargs)
            return result
        return wrapper
```

```python
p1 = function_call_count(name="Loss-Function", count=0)
loss_func=p1.call_func(cce2503_2022_loss_function)
p1.reset_count()

def get_initial_bracket(f, x=0, s=1e-2, k=2.0):
    """
    Find an interval [a,b] along a coordinate direction in which a minimum exists
    f : multivariate function where minimum exists (n-dimensional)
    x : starting position  (vector in n-dimensions)
    s : initial step size (scalar)
    k : interval expanding factor (scalar)
    returns:
    [a,b] : return scalar quantities a, b (bracket or interval limits)
    """
    a = x
    ya = f(x)
    b = a + s
    x=b
    yb = f(x) 
    if yb > ya:
        a, b = b, a
        ya, yb = yb, ya
        s = -s
    flag=True
    while flag:
        c = b + s
        x = c
        yc = f(x)
        if yc > yb:
            flag = False
        else:
            a, ya, b, yb = b, yb, c, yc # comment this line to preserve starting point
            s *= k
    if a < c: a_b = [a, c] 
    else: a_b=[c, a]
    return a_b
    
def interval_halving_algo(cost_func, a, b, e = 1e-15):
    """
    cost_func:  function that computes cost or loss
    [a,b] : Intial interval of uncertainty
    e : target interval of uncertainty
    max_itr: maximum number of iterations
    returns:
    [a,b]: Final interval of uncertainty
    n : number of iterations
    """
    #
    n=0
    D = b - a  #Interval of uncertainty
    D=D.mean()
    x_m = a + D/2
    y_m = cost_func(x_m)        
    #
    while D>e :
        n += 1
        x_l = a + D/4
        y_l = cost_func(x_l)
        x_u = b - D/4
        y_u = cost_func(x_u)
        if y_l < y_m:
            b = x_m
            x_m = x_l
            y_m = y_l
            D = b - a
            D=D.mean()
        elif y_u < y_m:
            a = x_m
            x_m = x_u
            y_m = y_u
            D = b - a
            D=D.mean()
        else:
            a = x_l
            b = x_u
            D = b - a
            D=D.mean()
    #
    return a, b, n


```

```python

    
#5 dimensional vector of zeros
z=np.zeros(5)




def coordinate_descent(cost_func, z0, e=1e-15):
    """
    cost_func:  function that computes cost or loss
    z0 : initial point
    e : target interval of uncertainty
    max_itr: maximum number of iterations
    returns:
    z_list: list of points visited
    """
    eroot=np.power(e,1.0/5.0)
   
    n=z0.size
    z_list = z0
      
     #minimize the cost function along every coordinate using interval  
    for i in range(n):
     
     #define lambda function to minimize along a coordinate
      g=lambda x: cost_func(np.insert(np.delete(z_list,i),i,x))  
      #get initial interval of uncertainty
      a_b=get_initial_bracket(g,z_list[i],s=1e-2,k=2.0)
      #minimize along the coordinate
      a,b,n=interval_halving_algo(g,a_b[0],a_b[1],e=eroot)
      #update the point
      z_list[i]=a
      z_list=np.insert(np.delete(z_list,i),i,b)
    return z_list
```
